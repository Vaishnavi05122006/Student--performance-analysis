```python
# STUDENT PERFORMANCE DATA ANALYSIS - Mobile Friendly Version (FIXED URL)
# Paste this ENTIRE block into ONE cell in Colab, then run it.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://github.com/YBI-Foundation/Dataset/raw/refs/heads/main/Students%20Performance.csv"
df = pd.read_csv(url)
print(df.head())

print("Dataset shape:", df.shape)
print(df.isnull().sum())

df['average_score'] = (df['math score'] + df['reading score'] + df['writing score']) / 3

plt.figure(figsize=(6,4))
sns.barplot(x='gender', y='average_score', data=df)
plt.title('Average Score by Gender')
plt.show()

plt.figure(figsize=(7,4))
sns.barplot(x='parental level of education', y='average_score', data=df)
plt.title('Average Score by Parental Education Level')
plt.xticks(rotation=45, ha='right')
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(x='test preparation course', y='average_score', data=df)
plt.title('Average Score by Test Prep Completion')
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(x='lunch', y='average_score', data=df)
plt.title('Average Score by Lunch Type')
plt.show()

correlation = df[['math score', 'reading score', 'writing score']].corr()
print("Correlation between subjects:")
print(correlation)

plt.figure(figsize=(6,4))
sns.heatmap(correlation, annot=True, cmap='Blues')
plt.title('Correlation Between Subject Scores')
plt.show()

print("Average score by test preparation:")
print(df.groupby('test preparation course')['average_score'].mean())

print("Average score by parental education:")
print(df.groupby('parental level of education')['average_score'].mean().sort_values(ascending=False))
```
